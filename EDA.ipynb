```python
# Check for null values
print(df.isnull().sum())

# Drop rows with missing delay values (you can choose to fill instead if needed)
df.dropna(subset=['carrier_delay', 'weather_delay', 'nas_delay', 'security_delay', 'late_aircraft_delay'], inplace=True)

# Convert year and month to datetime (if needed)
df['date'] = pd.to_datetime(df[['year', 'month']].assign(day=1))  # Assigning a dummy day
```

```python
delay_columns = ['carrier_delay', 'weather_delay', 'nas_delay', 'security_delay', 'late_aircraft_delay']
delay_sums = df[delay_columns].sum().sort_values(ascending=False)
print(delay_sums)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.barplot(x=delay_sums.index, y=delay_sums.values)
plt.title('Total Delay by Delay Type')
plt.ylabel('Total Delay (minutes)')
plt.xlabel('Delay Cause')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
carrier_delay = df.groupby('carrier_name')['carrier_delay'].sum().sort_values(ascending=False)
print(carrier_delay.head(10))

# Plot
carrier_delay.head(10).plot(kind='bar', figsize=(10, 6), title='Top 10 Carriers by Total Carrier Delay')
plt.ylabel('Total Delay (minutes)')
plt.xlabel('Carrier')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
airport_delay = df.groupby('airport_name')['arr_delay'].sum().sort_values(ascending=False)
print(airport_delay.head(10))

airport_delay.head(10).plot(kind='bar', figsize=(10, 6), title='Top 10 Airports by Total Arrival Delay')
plt.ylabel('Total Arrival Delay (minutes)')
plt.xlabel('Airport')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
monthly_delay = df.groupby('month')['carrier_delay'].mean()
monthly_delay.plot(kind='line', marker='o', figsize=(8, 5), title='Average Carrier Delay by Month')
plt.xlabel('Month')
plt.ylabel('Average Delay (minutes)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
pivot = df.pivot_table(values='carrier_delay', index='carrier_name', columns='month', aggfunc='mean')
print(pivot.head())
```

```python
df['delay_ratio'] = df['arr_del15'] / df['arr_flights']
```

```python
airport_traffic = df.groupby('airport')['arr_flights'].sum()
busy_airports = airport_traffic[airport_traffic > airport_traffic.quantile(0.75)].index
df['is_busy_airport'] = df['airport'].isin(busy_airports).astype(int)
```

```python
delay_columns = ['carrier_delay', 'weather_delay', 'nas_delay', 'security_delay', 'late_aircraft_delay']
df['total_delay'] = df[delay_columns].sum(axis=1)
```

```python
def get_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    else:
        return 'Fall'

df['season'] = df['month'].apply(get_season)
```

```python
df['day_of_week'] = df['date'].dt.day_name()
delay_by_day = df.groupby('day_of_week')['arr_delay'].mean().sort_values()
delay_by_day.plot(kind='bar', title='Average Arrival Delay by Day of Week')
plt.ylabel('Avg Delay (minutes)')
plt.show()
```

```python
major_airlines = ['Delta Air Lines Inc.', 'United Air Lines Inc.']
airline_compare = df[df['carrier_name'].isin(major_airlines)]
airline_group = airline_compare.groupby('carrier_name')[['carrier_delay', 'weather_delay', 'nas_delay']].mean()
airline_group.plot(kind='bar', stacked=True, title='Average Delay Causes by Airline')
plt.ylabel('Average Delay (minutes)')
plt.show()
```

```python
sns.scatterplot(data=df, x='weather_delay', y='nas_delay')
plt.title('Weather Delay vs NAS Delay')
plt.xlabel('Weather Delay (min)')
plt.ylabel('NAS Delay (min)')
plt.show()
```

```python
avg_delay_by_airline = df.groupby('carrier_name')['arr_delay'].mean().sort_values()
print(avg_delay_by_airline.head())  # Least average delay
```

```python
df['year'] = pd.to_datetime(df['date']).dt.year
yearly_delay = df.groupby('year')['arr_delay'].mean()
yearly_delay.plot(marker='o', title='Average Delay Over Years')
plt.ylabel('Avg Delay (min)')
plt.show()
```

```python
monthly_trend = df.groupby('date')[['carrier_delay', 'weather_delay', 'nas_delay']].mean()
monthly_trend.plot(title='Monthly Trend of Delay Types', figsize=(12, 6))
plt.ylabel('Average Delay (min)')
plt.show()
```
